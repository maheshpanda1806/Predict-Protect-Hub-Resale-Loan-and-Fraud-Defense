```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/spam.csv');
df.head(100)
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
Y = df['spam'].values
X = df.drop(columns='spam',axis=1).values
```

```python
#Calling the count vectorizer
from sklearn.feature_extraction.text import CountVectorizer

# Assuming your messages are stored in a text column named 'message' in the original DataFrame 'df'
# (replace 'message' with the actual column name if it's different)


# Create CountVectorizer with lowercase=True
vectorizer = CountVectorizer()

# Fit and transform (lowercase conversion happens here)
X_vectorized = vectorizer.fit_transform(df.Message)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_vectorized,Y,test_size = 0.2,random_state=0);
```

```python
#Traing the Model
from sklearn.naive_bayes import MultinomialNB
md = MultinomialNB()
md.fit(X_train,Y_train)
```

```python
#accuracy test
emails = [
    '']
emails_count = vectorizer.transform(emails)
ans = md.predict(emails_count)
if ans[0] == 0:
    print("Ham");
else:
       print("spam");
```

```python
md.score(X_test,Y_test)
```
