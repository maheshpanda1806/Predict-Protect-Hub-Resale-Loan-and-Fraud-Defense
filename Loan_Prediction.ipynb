```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loan_df = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv');
#loan_df.head()
loan_df.describe()
```

```python
loan_df.isnull().sum()
```

```python
loan_df = loan_df.dropna()
loan_df.isnull().sum()
loan_df.shape
loan_df.head(20)
```

```python
loan_df.groupby("Property_Area").describe()
```

```python
loan_df = loan_df.drop(['Loan_ID'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
loan_df['Gender'] = encoder.fit_transform(loan_df['Gender'])
loan_df['Married'] = encoder.fit_transform(loan_df['Married'])
loan_df['Education'] = encoder.fit_transform(loan_df['Education'])
loan_df['Self_Employed'] = encoder.fit_transform(loan_df['Self_Employed'])
loan_df['Property_Area'] = encoder.fit_transform(loan_df['Property_Area'])
loan_df['Loan_Status'] = encoder.fit_transform(loan_df['Loan_Status'])
loan_df = loan_df.replace(to_replace='3+', value=4)
```

```python
loan_df.head(20)
```

```python
loan_df['Dependents'] = loan_df['Dependents'].astype(int)
```

```python
loan_df["Dependents"].describe()
```

```python
Y = loan_df['Loan_Status']
X = loan_df.drop(['Loan_Status'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
# accuracy score on test data
X_test_prediction = classifier.predict(x_test)
test_data_accuray = accuracy_score(X_test_prediction,y_test)
print('Accuracy on testing data : ', test_data_accuray)
```
